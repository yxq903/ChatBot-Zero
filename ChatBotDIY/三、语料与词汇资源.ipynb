```python
import sys
import nltk
from nltk.corpus import brown
```

```python
genre_word = [(genre,word) for genre in brown.categories()
                           for word  in brown.words(categories=genre)]
```

```python
len(genre_word)
```

```python
genre_word[1161100]
```

```python
# 创建条件频率分布
cfd = nltk.ConditionalFreqDist(genre_word)
```

```python
# 指定条件和样本作图
cfd.plot(conditions=['news','adventure'], 
         samples=[u'stock', u'sunbonnet', u'Elevated', u'narcotic', u'four', u'woods',
                  u'railing', u'Until', u'aggression', u'marching', u'looking',
                  u'eligible', u'electricity', u'$25-a-plate', u'consulate',
                  u'Casey', u'all-county', u'Belgians', u'Western', u'1959-60',
                  u'Duhagon', u'sinking', u'1,119', u'co-operation', u'Famed', u'regional',
                  u'Charitable', u'appropriation', u'yellow', u'uncertain', u'Heights',
                  u'bringing', u'prize', u'Loen', u'Publique', u'wooden', u'Loeb', u'963', 
                  u'specialties', u'Sands', u'succession', u'Paul', u'Phyfe'])
```
